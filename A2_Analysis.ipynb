```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
df=pd.read_json('stat.json')
```

```python
df['Heuristic Number']=df['heur_name'].str.split('_',expand=True)[1]
df['heur_short_name'] = df['heur_name'].str.split('_',expand=True)[0]
df.columns
df=df.round(3)
file_count=50
```

```python
#Average Sol Path
plt.figure(figsize=(12, 8))
ax=sns.barplot(data=df,x="heur_name", y="av_sol_length", palette="colorblind")
plt.xticks(rotation=45)
plt.title("Solutions Paths Average Lengths")
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1, height,
            ha='center', va='bottom')
plt.ylabel('Length')
plt.xlabel('')
plt.savefig('Analysis_figures/3x3/average_sol_path_length.png')
```

```python
#Total Sol Path
plt.figure(figsize=(12, 8))
ax=sns.barplot(data=df,x="heur_name", y="total_sol_length", palette="colorblind")
plt.xticks(rotation=45)
plt.title("Solutions Paths Total Lengths")
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1, height,
            ha='center', va='bottom')
plt.ylabel('Length')
plt.xlabel('')
plt.savefig('Analysis_figures/3x3/total_sol_path_length.png')
```

```python
#Average Execution time
plt.figure(figsize=(12, 8))
ax=sns.barplot(data=df,x="heur_name", y="av_execution_time", palette="colorblind")
plt.xticks(rotation=45)
plt.title("Average Execution Time for 3x3")
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1, height,
            ha='center', va='bottom')
plt.ylabel('Time (s)')
plt.xlabel('')
plt.savefig('Analysis_figures/3x3/average_exec_time.png')
```

```python
#Total Execution time
plt.figure(figsize=(12, 8))
ax=sns.barplot(data=df,x="heur_name", y="total_execution_time", palette="colorblind")
plt.xticks(rotation=45)
plt.title("Total Execution Time")
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1, height,
            ha='center', va='bottom')
plt.ylabel('Time (s)')
plt.xlabel('')
plt.savefig('Analysis_figures/3x3/total_exec_time.png')
```

```python
#Average Cost
plt.figure(figsize=(12, 8))
ax=sns.barplot(data=df,x="heur_name", y="av_cost", palette="colorblind")
plt.xticks(rotation=45)
plt.title("Average Solution Cost for 3x3")
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1, height,
            ha='center', va='bottom')
plt.ylabel('Cost')
plt.xlabel('')
plt.savefig('Analysis_figures/3x3/average_solution_cost.png')
```

```python
sub_df=df[['heur_name','n_solved','n_unsolved']]
sub_df=sub_df.append(sub_df,ignore_index=True)
sub_df['solved']=pd.Series([True,True,True,True,True,False,False,False,False,False])
sub_df['n']=sub_df['n_solved'][sub_df.solved==True]
sub_df['n'][sub_df.solved==False]=sub_df['n_unsolved'][sub_df.solved==False]
sub_df
```

```python
#Success Rate
plt.figure(figsize=(12, 8))

ax=sns.barplot(data=sub_df,x="heur_name", y="n",hue='solved',hue_order=[True,False], palette="colorblind")
plt.xticks(rotation=45)
plt.title("Puzzles Solved for 3x3")
for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1,'{}, ({} %)'.format(int(height),height*100/file_count),
            ha='center', va='bottom',size=12)
    plt.savefig('Analysis_figures/3x3/success.png')
    
plt.ylabel('# of puzzles')
plt.xlabel('')
plt.savefig('Analysis_figures/3x3/success.png')
```
